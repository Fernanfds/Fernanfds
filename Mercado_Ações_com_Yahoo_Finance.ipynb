```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import datetime as dt
import pandas_datareader.data as web

!pip install yfinance --upgrade --no-cache-dir
import yfinance as yf
yf.pdr_override()

```

```python
ibov = web.get_data_yahoo('PETR4.SA', start='2020-06-01', end='2023-06-01')
```

```python
ibov.head()
```

```python
# @title High

from matplotlib import pyplot as plt
ibov['High'].plot(kind='line', figsize=(8, 4), title='High')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
ibov.tail()
```

```python
ibov["Close"].plot(figsize=(16,8), label="PETR4.SA")
ibov["Close"].rolling(21).mean().plot(label="MM21")
ibov["Close"].rolling(50).mean().plot(label="MM50")
plt.legend()
```

```python
ibov_fatiado = ibov[ibov.index.year == 2023]
ibov_fatiado["Close"].plot(figsize=(22,8), label="PETR4.SA")
ibov_fatiado["Close"].rolling(21).mean().plot(label="MM21")
ibov_fatiado["Close"].rolling(200).mean().plot(label="MM200")
plt.legend()
```
