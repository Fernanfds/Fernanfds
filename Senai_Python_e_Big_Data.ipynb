```python
import pandas as pd

data = {
    'Nome': ['João', 'Maria', 'Pedro', 'Ana'],
    'idade': [30, 25, 35, 19],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Campinas'],
    'Profissão': ['Engenheiro', 'Médico', 'Advogado', 'Feirante'],
    'Salário': [5000, 6000, 7000, 2000]
}

df = pd.DataFrame(data)

df
```

```python
genero = ['M', 'F', 'M', 'F']
df['Genero'] = genero
df
```

```python
endereco = ['Rua A', 'Rua B', 'Rua C', 'Rua D']
df.insert(1, 'Endereço', endereco)
df
```
