```python
acoes = ['BBAS3.SA', 'PETR4.SA', 'VALE3.SA', 'CMIG4.SA']
```

```python
dados_acoes = yf.download(acoes, start='2023-01-01', end='2024-06-01')
```

```python
dados_acoes['Close'].plot()
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.title('Rastreador de Ações')
plt.legend(acoes, fontsize=7)
plt.show()
```

```python
dados_normalizados = dados_acoes['Close'] / dados_acoes['Close'].iloc[0]
dados_normalizados.plot()
plt.xlabel('Data')
plt.ylabel('Preço Normalizado')
plt.title('Rastreador de Ações')
plt.legend(acoes, fontsize=7)
plt.show()
```

Geração de Números Aleatórios

```python
import matplotlib.pyplot as fig
import random
aleat=[]
for i in range(1000):
    x=random.uniform(-1,1)
    aleat.append(x)

fig.subplot(211)
fig.plot(aleat,color='black')
fig.xlabel('Eixo Horizontal')
fig.ylabel('Números Aleatórios Entre -1 e 1')

fig.subplot(212)
fig.hist(aleat,bins=20,color='gray')
fig.xlabel('Classes')
fig.ylabel('frequência')

fig.show()
```

```python
import matplotlib.pyplot as fig
import random
from scipy.stats import norm
import statistics as st
import numpy as ny

aleat=[]
for i in range(1000):
    x=random.gauss(0,1)
    aleat.append(x)

fig.subplot(211)
fig.plot(aleat,color='black')
fig.xlabel('Eixo Horizontal')
fig.ylabel('Números Aleatórios Entre -1 e 1')

fig.subplot(212)
fig.hist(aleat,bins=20,color='gray', density=True)
fig.xlabel('Classes')
fig.ylabel('frequência')

xmin,xmax = fig.xlim()
media=st.mean(aleat)
desvio=st.pstdev(aleat)
eixo_x=ny.linspace(xmin,xmax,100)
eixo_y=norm.pdf(eixo_x,media,desvio)
fig.plot(eixo_x,eixo_y,color='black')

fig.show()
```
